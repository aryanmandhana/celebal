```python
df = df[df["LotFrontage"].notna()]
```

```python
df.isna().sum()
```

```python
q1 = df["LotFrontage"].quantile(0.25)
q3 = df["LotFrontage"].quantile(0.75)

xyz = q3 - q1

lower_range = q1 - (1.5 * xyz)
upper_range = q3 + (1.5 * xyz)

outliers = len(df[(df["LotFrontage"] < lower_range) | (df["LotFrontage"] > upper_range)])
print("Number of LotFrontage having Outliers : ", outliers)
```

```python
missing_values = data.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
for col in ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtExposure', 'BsmtFinType2', 'BsmtFinType1', 'BsmtCond', 'BsmtQual', 'MasVnrType']:
    data[col] = data[col].fillna('None')

for col in ['GarageYrBlt', 'MasVnrArea']:
    data[col] = data[col].fillna(0)

data['LotFrontage'] = data['LotFrontage'].fillna(data['LotFrontage'].median())

data['Electrical'] = data['Electrical'].fillna(data['Electrical'].mode()[0])
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder

def label_encode(columns):
    for col in columns:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
    return data

ordinal_cols = ['ExterQual', 'ExterCond', 'HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 
                'GarageCond', 'PoolQC', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                'BsmtFinType2', 'Functional', 'PavedDrive']

data = label_encode(ordinal_cols)
```

```python
data = pd.get_dummies(data, columns=[col for col in categorical_cols if col not in ordinal_cols])
```
